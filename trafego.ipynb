#1.7 Gráficos

O gráfico abaixo nos demonstra se houve um aumento ou diminuição no tráfego entre os anos 1993-2019.

```python
import matplotlib.pyplot as plt

media_por_ano = (
    data.groupby("year")["all_motor_vehicles"]
    .mean()
)

plt.figure(figsize=(10,5))
plt.plot(
    media_por_ano.index,
    media_por_ano.values
)

plt.title("Média de veículos por ano")
plt.xlabel("Ano")
plt.ylabel("Média de veículos")
plt.grid()

plt.show()
```

Distribuição do volume de veículos.

```python
plt.hist(data["all_motor_vehicles"])
```

Abaixo temos mais um exemplo de gráfico, dessa vez, iremos ver as 10 regiões do DataFrame com maior tráfego.

```python
top10 = (
    data.groupby("local_authority_name")
    ["all_motor_vehicles"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,5))
top10.plot(kind="bar")

plt.title("Top 10 regiões com maior tráfego")
plt.xlabel("Região")
plt.ylabel("Volume de veículos")

plt.show()
```
